### Loading Data <a class="anchor" id="section1_1"></a>
First, the necessary libraries are loaded into the notebook. The pandas library is used to import data from Drug_overdose_py.csv and preview the first five rows of the DataFrame. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Drug_overdose_py.csv")
df.head()
```

### Data Information / Cleaning <a class="anchor" id="section1_2"></a>

Initial insights:

* 15 Columns and 6228 Rows
* There are missing data points in the `ESTIMATE` & `FLAG` Columns
* Some columns have a corresponding numbering system ie. (`PANEL` & `PANEL_NUM`)
* The focus of this data set will be the `ESTIMATE` column in regard to the `PANEL` / `PANEL_NUM` & `YEAR` columns; values are floats and integers 
* Column names need to be modified for ease of use

```python
#Lower case all column names for easier manimpulation of the table
df.columns = [x.lower() for x in df.columns]
df.head()
```

```python
#Checking the qulities of the table and how each column corresponds with eachother 
print(df.dtypes)
print("\n(ROWS,COLS):",df.shape)

print("\nYEAR:",df.year.unique())
print("YEAR NUM:",df.year_num.unique())
print("Number of YEARS:",df.year.nunique())

print("\nPANEL TYPE:",df.panel.unique())
print("PANEL NUM:",df.panel_num.unique())
print("Number of PANEL TYPE:",df.panel.nunique())

print("\nUNIT:",df.unit.unique())
print("UNIT NUM:",df.unit_num.unique())
print("Number of UNITS:",df.unit.nunique())

print("\nSTUB NAME:",df.stub_name.unique())
print("STUB NAME NUM:",df.stub_name_num.unique())
print("Number of STUB NAMES:",df.stub_name.nunique())

print("\nSTUB LABEL:",df.stub_label.unique())
print("STUB LABEL NUM:",df.stub_label_num.unique())
print("Number of STUB LABELS:",df.stub_label.nunique())

print("\nAGE:",df.age.unique())
print("AGE NUM:",df.age_num.unique())
print("Number of AGES:",df.age.nunique())
```

```python
#Checking for missing values
print(df.isnull().sum())
print()

#Showing where the missing values are coming from
missing_values_by_panel = df.groupby(["panel_num", "unit_num", "stub_name"]).apply(lambda x: x.isnull().sum())
print(missing_values_by_panel)
print()

#panel_num_value = 3
#missing_values_for_panel_num = missing_values_by_panel.loc[panel_num_value]
#print(f"Missing values for panel_num {panel_num_value}:\n{missing_values_for_panel_num}")
```

## Reorganization <a class="anchor" id="reorganization"></a>

While exploring the data set it proved that a restructuring of the data was necessary to begin analysis:

* Create New DataFrames:
    1. Male subjects by `panel_num` 0, which corresponds to `panel` - "All drug overdose deaths"
    2. Female subjects by `panel_num` 0, which corresponds to `panel` - "All drug overdose deaths"
    3. Age Groups subjects by `panel_num` 0, which corresponds to `panel` - "All drug overdose deaths"

```python
dfm_all = df[(df.stub_label == "Male") & (df.panel_num == 0) & (df.unit_num == 1) ].reset_index(drop=True)
dfm_all.head()
```

```python
dff_all = df[(df.stub_label == "Female") & (df.panel_num == 0) & (df.unit_num == 1) ].reset_index(drop=True)
dff_all.head()
```

```python
df_age_group_avg = df[(df.panel_num == 0) & (df.stub_name == "Age")].groupby(["age"]).estimate.mean().reset_index()
#print(df_age_group_avg)

# Move target row to first element of list.
target_row = 8
idx = [target_row] + [i for i in range(len(df_age_group_avg)) if i != target_row]
df_age_group_avg = df_age_group_avg.iloc[idx].reset_index(drop=True)
print("Average Estimate:\n",df_age_group_avg)

df_age_group_median = df[(df.panel_num == 0) & (df.stub_name == "Age")].groupby(["age"]).estimate.median().reset_index()

# Move target row to first element of list.
target_row = 8
idx = [target_row] + [i for i in range(len(df_age_group_median)) if i != target_row]
df_age_group_median = df_age_group_median.iloc[idx].reset_index(drop=True)
print()
print("Median Estimate:\n",df_age_group_median)

df_age_group = df.pivot_table("estimate", index="year", columns="age").drop("All ages", axis=1)
df_age_group = df_age_group[['Under 15 years'] + [col for col in df_age_group.columns if col != 'Under 15 years' ]].reset_index()
df_age_group.head()
```

## Exploratory Data Analysis <a class="anchor" id="eda"></a>

The following independent variables will be the focus of this analysis:
* `Sex`
* `Age`

The goal will be to see how each of these are associated with drug overdose death rates which is the dependant variable `estimate`

### Male / Female <a class="anchor" id="section3_1"></a>

This section is the analysis of all drug overdose deaths with the Male and Female populations as a whole. The range of time spans 20 years, 1999-2018. It suggests that the death rate for men has been more parabolic and at a higher rate compared to their female counterparts, while the female population has seen a more linear progression over time. 

```python
plt.plot(dfm_all.year, dfm_all.estimate, marker="o")
plt.plot(dff_all.year, dff_all.estimate, c="orange", marker="o")
plt.xticks(range(2000,2019,2))
plt.legend(["Male", "Female"])
plt.xlabel("Year")
plt.ylabel("Deaths per 100,000 Resident Population", fontsize=10)
plt.title("All Drug Overdose Deaths (1999-2018)")
plt.show()
plt.clf()
```

#### Closer Look <a class="anchor" id="section3_1.1"></a>

A closer look is taken to inspect each population individually since `male` drug overdose death rates are much greater than that of `female` drug overdose death rates. This could skew the interpretation of the `female` population.

As seen below, the `female` drug overdose death rate was indeed skewed compared to the scale of the `male` drug overdose death rates. The acceleration of the `female` drug overdose death rate has actually been more parabolic in relation to their `male` counterparts, although more males in absolute terms lose their lives to overdose. In the 20 years from 1999-2018 males have experienced a "cool-off" period twice while females have only experienced one. During the first "coo-off" period for the `male` population (2006-2010) the `female` drug overdose death rate continued to rise. When the second "cool-off" period for the `male` population (2011-2012) occurred the `female` population followed suit. 

```python
plt.figure(figsize=(8,6))

ax1 = plt.subplot(1,2,1)
ax1.plot(dfm_all.year, dfm_all.estimate, marker="o", label="Male")
plt.xlabel("Year")
plt.ylabel("Deaths per 100,000 Resident Population")
ax1.legend()


ax2 = plt.subplot(1,2,2)
ax2.plot(dff_all.year, dff_all.estimate, c="orange", marker="o", label="Female")
plt.xlabel("Year")
plt.ylabel("Deaths per 100,000 Resident Population")
ax2.legend()

plt.suptitle("Drug Overdose Deaths (1999-2018)", y=0.93)
plt.subplots_adjust(wspace=0.3)
plt.show()
```

### Age Groups <a class="anchor" id="section3_2"></a>

Here, a look is taken at the average death rate from 1999-2018 for each age category detailed below. The charts indicate that the age group of `45-54` year olds have been the most affected, closely followed by the `35-44` age group. The `Under 15` group could be considered an outlier with extreme differences amoungst the rest of the age categories.

```python
plt.figure(figsize=(8,6))
color = ["blue", "orange", "green", "red", "purple", "brown", "yellow", "magenta", "gray"]
age_cats = df_age_group_avg.age.unique()
bar = plt.bar(df_age_group_avg.age, df_age_group_avg.estimate)
for i in range(len(bar)):
    if i < len(color):
        bar[i].set_color(color[i])
plt.xticks(age_cats, labels=["Under 15", "15-24", "25-34", "35-44", "45-54", "55-64", "65-74", "75-84", "85+"])
plt.title("Average Drug Overdose Deaths (1999-2018)")
plt.xlabel("Age Group")
plt.ylabel("Deaths per 100,000 Resident Population")
plt.show()
plt.clf()
```

```python
colors = ["blue", "orange", "green", "red", "purple", "brown", "yellow", "magenta", "gray"]
cats = ["Under 15", "15-24", "25-34", "35-44", "45-54", "55-64", "65-74", "75-84", "85+"]
plt.figure(figsize=(8,6))
plt.pie(df_age_group_avg.estimate, labels=cats, autopct="%1.1f%%", colors=colors)
plt.legend(bbox_to_anchor=(1.2,1.05), title="Age Group")
plt.axis("equal")
plt.title("Average Drug Overdose Deaths (1999-2018)\n(per 100,000 Resident Population)")
plt.show()
```

#### Breakdown (Range) <a class="anchor" id="section3_2.1"></a>

Below is the range for the estimated death rate for each category. The biggest difference belongs to the `25-34` age group with a difference of almost 15. The next biggest differnece belongs to the `35-44` age group with a difference of almost 13. Once again the `Under 15` age group ccould be considered an outlier with a difference of only 0.04.

Further down the age groups are individually broken down to see the changes over time (1999-2018)

```python
for i in df_age_group.columns:
    if i != "year":
        print(f"\n1999-2018:\nRange of Estimate ({i}): {round(min(df_age_group[i]),2)} - {round(max(df_age_group[i]),2)} \nDifference of: {round(max(df_age_group[i])-min(df_age_group[i]),2)}")
```

#### Breakdown (Line Charts) <a class="anchor" id="section3_2.2"></a>

As shown below in the first set of line charts are the age groups individually plotted to see drug overdose deaths over time (1999-2018). Out of the 9 age groups there are only two that do not follow a parabolic form. Those two age groups are `Under 15` and `75-84` respectivly. The other "out of ordinary" age group in relation to this data set is the `85+` age group, although the trend is still parabolic like the others it just trends in the opposite direction. 

The second set of line charts illustrates drug overdose deaths for the next 20 years (2019-2038) if the trend continues as it has in the past 20 years (1999-2018). All but one age group, `85+`, is projected to increase into the future if the trend continues.  

```python
count = 1
plt.figure(figsize=(16,16))
color = ["blue", "orange", "green", "red", "purple", "brown", "yellow", "magenta", "gray"]
for i, col in enumerate(df_age_group.columns):
        if col != "year":
            plt.subplot(3,3, count)
            count += 1 
            plt.plot(df_age_group.year, df_age_group[col], marker="o", label= f"{col}", color=color[count- 2])
            plt.xlabel("Year")
            plt.ylabel("Deaths per 100,000 Resident Population")
            plt.legend()

plt.suptitle("Drug Overdose Deaths by Age Group (1999-2018)", y=0.93, fontsize= 16)
plt.show()
plt.clf()
```

```python
from scipy.stats import linregress

years = df_age_group.year.unique()
future_years = np.arange(min(years), max(years) + 21)  # Projecting 20 years into the future

count = 1
plt.figure(figsize=(16, 16))
color = ["blue", "orange", "green", "red", "purple", "brown", "yellow", "pink", "gray"]

for i, col in enumerate(df_age_group.columns):
    if col != "year":
        plt.subplot(3, 3, count)
        count += 1
        plt.plot(df_age_group.year, df_age_group[col], marker="o", label=f"{col}", color=color[count - 2])

        # Linear regression for existing data
        slope, intercept, _, _, _ = linregress(df_age_group.year, df_age_group[col])
        regression_line = slope * df_age_group.year + intercept
        plt.plot(df_age_group.year, regression_line, color="black", linestyle="-", label="Trend")

        # Projecting linear regression line for the next 20 years
        future_regression_line = slope * future_years + intercept
        plt.plot(future_years, future_regression_line, color="black", linestyle="--", alpha=0.5, label="Continuing Trend")

        plt.xlabel("Year")
        plt.ylabel("Deaths per 100,000 Resident Population")
        plt.legend()

plt.suptitle("Drug Overdose Deaths by Age Group (1999-2038)", y=0.93, fontsize=16)
plt.show()
plt.clf()
```

### Conclusion <a class="anchor" id="conclusion"></a>

#### Findings Overview

From the information included in the data set analyzed, there is a big problem in the United States and its drug culture. The `Male` population has a higher chance of dying from a drug overdose than the `female` population, although both are concerning issues in their own right. Between the age groups `85+` is the only one that is in a down trend while all the others are set to continue upward. In the case for the `Under 15` age group the projection shows for it to increase but only at a slight rate compared to the others.
